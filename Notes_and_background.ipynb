## Synthetic AID-like (Activation-Induced Deaminase) Mutation Generator

### Model Overview

* **Hidden states encode binding only**:

  * `0` = Unbound / background
  * `1` = Bound / scanning
* **“Firing”** is an emission event that occurs **only in state 1**, with a **context-dependent probability**.

---

### Context-Dependent Firing Bias

Firing probability is modulated by local sequence context using bias multipliers:

* **`hotspot_bias`** — increases firing probability
* **`coldspot_bias`** — decreases firing probability

These biases depend on well-characterized AID sequence motifs.

---

### Hotspot Motif: WRC

**Pattern:**

```
[A/T] [A/G] C
```

| Symbol | Meaning                  |
| ------ | ------------------------ |
| **W**  | A or T (weak base pairs) |
| **R**  | A or G (purine)          |
| **C**  | Target cytosine          |

* **Effect:** AID fires **10× more often**
* **Bias:** `hotspot_bias = 10.0`
* **Examples:** `AAC`, `TAC`, `AGC`, `TGC`

---

### Coldspot Motif: SYC

**Pattern:**

```
[G/C] [C/T] C
```

| Symbol | Meaning                    |
| ------ | -------------------------- |
| **S**  | G or C (strong base pairs) |
| **Y**  | C or T (pyrimidine)        |
| **C**  | Target cytosine            |

* **Effect:** AID fires **0.1× as often**
* **Bias:** `coldspot_bias = 0.1`
* **Examples:** `GCC`, `GTC`, `CCC`, `CTC`

---

### Mutation Formation

* Mutations occur **only at cytosines (C)**

* Probability of mutation:

  ```
  p_mut = p_fire × mutation_eff
  ```

* **`mutation_eff`**: Repair failure rate

  * ~80% of deaminated bases become fixed mutations

---

### Background Mutations

* **State 0 (unbound/background)** allows low-level noise
* Background mutations occur with probability:

  ```
  p_noise = noise
  ```

### Biological Notes

#### The Chain of Roles

B cell
→ a specialized immune cell, a white blood cell specialized to produce one unique antibody.

Antibody
→ the protein the B cell makes to recognize threats, a protein that binds a specific molecular target.

Enzyme (AID)
→ a protein inside the B cell that modifies DNA to improve antibodies. Enzyme: protein that catalyzes a specific chemical reaction.
AID: An enzyme in B cells that deliberately mutates antibody DNA.

So:

B cells use enzymes to mutate antibody genes in order to make better antibodies.

#### Why “Base Pair” (bp)?

DNA is a double helix composed of two complementary strands. Bases pair according to fixed rules:

* **A–T** → 2 hydrogen bonds
* **C–G** → 3 hydrogen bonds

---

#### Why These Names?

**Weak vs. Strong**

Weak vs Strong refers to hydrogen bonding strength.

* **A–T** pairs have fewer hydrogen bonds → *weaker*
* **G–C** pairs have more hydrogen bonds → *stronger*

**Purines vs. Pyrimidines**

Purine vs Pyrimidine is chemical classification.

* **Purines:** A, G (two-ring molecules)
* **Pyrimidines:** C, T (single-ring molecules)
