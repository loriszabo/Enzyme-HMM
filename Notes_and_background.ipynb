## Synthetic AID-like (Activation-Induced Deaminase) Mutation Generator

### Model Overview

* **Hidden states encode binding only**:

  * `0` = Unbound / background
  * `1` = Bound / scanning
* **“Firing”** is an emission event that occurs **only in state 1**, with a **context-dependent probability**.

---

### Context-Dependent Firing Bias

Firing probability is modulated by local sequence context using bias multipliers:

* **`hotspot_bias`** — increases firing probability
* **`coldspot_bias`** — decreases firing probability

These biases depend on well-characterized AID sequence motifs.

---

### Hotspot Motif: WRC

**Pattern:**

```
[A/T] [A/G] C
```

| Symbol | Meaning                  |
| ------ | ------------------------ |
| **W**  | A or T (weak base pairs) |
| **R**  | A or G (purine)          |
| **C**  | Target cytosine          |

* **Effect:** AID fires **10× more often**
* **Bias:** `hotspot_bias = 10.0`
* **Examples:** `AAC`, `TAC`, `AGC`, `TGC`

---

### Coldspot Motif: SYC

**Pattern:**

```
[G/C] [C/T] C
```

| Symbol | Meaning                    |
| ------ | -------------------------- |
| **S**  | G or C (strong base pairs) |
| **Y**  | C or T (pyrimidine)        |
| **C**  | Target cytosine            |

* **Effect:** AID fires **0.1× as often**
* **Bias:** `coldspot_bias = 0.1`
* **Examples:** `GCC`, `GTC`, `CCC`, `CTC`

---

### Mutation Formation

* Mutations occur **only at cytosines (C)**

* Probability of mutation:

  ```
  p_mut = p_fire × mutation_eff
  ```

* **`mutation_eff`**: Repair failure rate

  * ~80% of deaminated bases become fixed mutations

---

### Background Mutations

* **State 0 (unbound/background)** allows low-level noise
* Background mutations occur with probability:

  ```
  p_noise = noise
  ```

## 1. The Problem AID Solves (Why It Exists at All)

Vertebrates face a fundamental challenge:

> **How do you generate a near-infinite diversity of antibodies using a finite genome—fast, and after birth?**

Random mutation is normally dangerous. But for adaptive immunity, **controlled, targeted mutation** is extremely useful.

AID exists because **B cells need a way to deliberately break the rules of genomic stability**, *locally and temporarily*, to improve antibodies.


---

## 2. What AID Actually Does (One Simple Reaction)

At its core, AID performs a single chemical reaction:

```
Cytosine (C) → Uracil (U)
```

But it does so:

* In **DNA**, not RNA
* On **single-stranded DNA**
* During **active transcription**

This creates a **U:G mismatch**, which DNA repair systems must deal with.

---

## 3. Turning Damage into Diversity

Once AID creates U in DNA, the cell has options:

### 1. Faithful Repair (No Mutation)

* Uracil is removed
* Cytosine restored
* No diversity

### 2. Error-Prone Repair (Desired Outcome)

Different repair pathways are intentionally exploited:

| Pathway                    | Outcome                    |
| -------------------------- | -------------------------- |
| Base excision repair       | Point mutations            |
| Mismatch repair            | Mutation clusters          |
| Double-strand break repair | Class switch recombination |

**AID doesn’t create mutations directly.**
It creates **lesions** that the cell *chooses* to repair inaccurately.

---

## 4. Spatial Targeting: Why Antibodies and Not the Whole Genome?

AID is extraordinarily dangerous. Unchecked, it causes cancer.

So evolution layered multiple safeguards:

### Targeting Rules

AID activity is enriched where:

* DNA is **single-stranded**
* Transcription is **high**
* Specific sequence motifs (WRC hotspots) occur
* Specialized cofactors recruit or stabilize AID

### Result

* Immunoglobulin loci get **hammered**
* Most of the genome is **mostly spared**
* “Mostly” is important—leakage causes lymphoma

---

## 5. Temporal Control: When AID Is Allowed to Exist

AID is:

* Expressed **only in activated B cells**
* Confined largely to the **nucleus only briefly**
* Rapidly degraded

This ensures:

> Mutation happens **after antigen exposure**, not during development or in stem cells.



# 1. What Are Immunoglobulin Loci?

### Start with “locus”

A **locus** is simply:

> **A specific location on a chromosome where a gene (or group of genes) lives.**

Your DNA is organized into chromosomes, and each chromosome has many loci.

---

### What “Immunoglobulin” Means

**Immunoglobulin (Ig)** is the formal name for:

> **Antibodies**

So:

> **Immunoglobulin loci = the regions of DNA that contain antibody genes.**

---

### Why There Are Multiple Immunoglobulin Loci

Antibodies are built from parts:

* **Heavy chain**
* **Light chain**

These parts live in different loci:

| Locus   | Encodes         |
| ------- | --------------- |
| **IGH** | Heavy chain     |
| **IGK** | Light chain (κ) |
| **IGL** | Light chain (λ) |

Each locus contains:

* Many gene segments (V, D, J, C)
* Switch regions (for class switching)
* Regulatory elements

These loci are **special**:

* They are deliberately rearranged
* They are deliberately mutated
* Nowhere else in the genome is treated this way

---

### Why AID Targets Immunoglobulin Loci

Immunoglobulin loci:

* Are highly transcribed in activated B cells
* Form single-stranded DNA during transcription
* Contain hotspot motifs (WRC)
* Have cofactors that recruit AID

This makes them **preferred targets** for mutation.

---

# 2. DNA Damage vs DNA Repair (Big Idea)

AID does **damage**, not mutation:

```
C → U
```

What happens next depends on **repair pathways**.

Different repair pathways → different biological outcomes.

---

# 3. Base Excision Repair (BER) → Point Mutations

### What BER Normally Does

BER fixes small, simple lesions (like uracil in DNA).

Steps (simplified):

1. Enzyme removes the damaged base
2. A single-base gap remains
3. DNA polymerase fills the gap
4. DNA ligase seals it

---

### How BER Creates Point Mutations in B Cells

In AID-targeted regions:

* Error-prone polymerases are used
* Wrong base is sometimes inserted

Result:

> **Single-base substitutions**

These are **point mutations**.

---

### Why This Matters

* Small changes → small shape changes
* Fine-tunes antibody binding
* Dominant mechanism in somatic hypermutation

---

# 4. Mismatch Repair (MMR) → Mutation Clusters

### What MMR Normally Does

Mismatch repair fixes:

* Base mismatches
* Small insertion/deletion errors

It usually improves fidelity.

---

### How MMR Creates Mutation Clusters in B Cells

In immunoglobulin loci:

1. U:G mismatch is recognized
2. A stretch of DNA around it is removed
3. Error-prone polymerases refill the region

Result:

> **Multiple mutations close together**

These are **mutation clusters**.

---

### Why Clustering Is Useful

* Changes multiple amino acids at once
* Allows larger jumps in antibody shape
* Accelerates evolution

---

# 5. Double-Strand Break Repair → Class Switch Recombination

### What a Double-Strand Break Is

* Both DNA strands are cut
* Extremely dangerous damage
* Usually avoided at all costs

---

### How AID Triggers Double-Strand Breaks

* AID deaminates many cytosines in **switch regions**
* Repair enzymes cut both strands
* Breaks accumulate

---

### What Class Switch Recombination (CSR) Is

CSR changes the **antibody’s constant region**, not its binding site.

Result:

* Same antigen specificity
* Different immune function

Examples:

* IgM → IgG (better opsonization)
* IgA (mucosal immunity)
* IgE (allergy / parasites)

---

### Mechanism Summary

| Repair Type                | Damage Outcome           | Biological Result |
| -------------------------- | ------------------------ | ----------------- |
| Base excision repair       | Single-base changes      | Point mutations   |
| Mismatch repair            | Multi-base changes       | Mutation clusters |
| Double-strand break repair | Large DNA rearrangements | Class switching   |

---

# 6. Why Different Outcomes Are Useful

The immune system needs:

* **Fine tuning** (point mutations)
* **Occasional big jumps** (clusters)
* **Functional switching** (CSR)

AID + repair pathways provide all three.

---

# 7. The Big Picture Chain (End-to-End)

1. Immunoglobulin loci encode antibodies
2. Activated B cells express AID
3. AID creates U lesions in Ig DNA
4. Repair pathways respond differently
5. Outcomes:

   * Better binding
   * Faster evolution
   * Functional diversity

# On The HMM inference principles

Below is a “read-the-code like a machine” walkthrough, with each piece tied explicitly to **HMM concepts** (states, transitions, emissions, forward algorithm, and MLE parameter fitting). I’ll keep the mapping consistent with your AID story: *unbound ↔ background noise* and *bound ↔ AID scanning with context-dependent firing*.

---

## 0) What this script does (big picture)

You have a dataset of paired sequences:

* `original_seq` (reference DNA)
* `mutated_seq` (observed output after simulated AID-like mutagenesis)

The script defines a **2-state Hidden Markov Model** for “where AID is” along the sequence:

* hidden state `0`: **unbound/background**
* hidden state `1`: **bound/scanning**

Given parameters θ =
`(alpha_recruit, lambda_scan, base_activation, hotspot_bias, coldspot_bias, mutation_eff)`

it computes:

[
\log P(\text{mutated_seq} \mid \text{original_seq}, \theta)
]

using the **forward algorithm** (in log-space), sums this over all sequences, and then uses numerical optimization to find θ that **maximizes likelihood** (equivalently minimizes negative log-likelihood).

This is standard **HMM maximum likelihood** (but not Baum–Welch / EM; it’s direct optimization using forward likelihood only).

## 1) Imports and constants

* Important modeling choice: **noise is treated as known**, not estimated. That reduces parameter coupling and makes optimization easier.

HMM interpretation: state 0 has an emission distribution that is “almost always copy original base, rarely substitute uniformly”.

## 2) `_clamp01(x)`

```py
def _clamp01(x: float) -> float:
    return float(max(0.0, min(1.0, x)))
```

This forces probabilities to stay in `[0,1]`. It’s used for things like `base_activation * hotspot_bias`, which could exceed 1.

HMM note: this is a *nonlinear saturation* in the emission model. It makes the model piecewise and can create flat regions in optimization (still fine, but good to know).

This function encodes your biology-inspired rule:

* context is the previous two bases: positions `t-2` and `t-1`

**HMM interpretation:**

## 5) Emission models in log-space

### Background state emission: `_log_emit_bg`

```py
if obs == orig: p = 1 - noise
else: p = noise / 3
return log(p)
```

This is a categorical distribution over {A,C,G,T} conditioned on orig:

* correct base with prob `1-noise`
* any of the other 3 bases with equal prob `noise/3`

**HMM mapping:**
[
P(O_t \mid S_t=0, \text{orig}_t) =
\begin{cases}
1-\epsilon & O_t = \text{orig}_t \
\epsilon/3 & O_t \neq \text{orig}_t
\end{cases}
]

### Bound state emission: `_log_emit_bound`

This is the key “AID-like” emission:

1. compute `p_fire(t)` from context
2. compute `p_mut(t) = p_fire(t) * mutation_eff`
3. allow only:

* “no mutation” (`obs == orig`) with prob `1 - p_mut`
* “C→T” (`orig == C and obs == T`) with prob `p_mut`
* anything else has prob 0

So in state 1, the model says:

* AID can only cause **C→T** mutations
* it only acts when reference base is C
* it never creates A/G changes directly (they’re impossible under this model)

**HMM mapping:**
[
P(O_t \mid S_t=1, \text{orig}, \text{context}) =
\begin{cases}
1-p_{\text{mut}}(t) & O_t=\text{orig}*t \
p*{\text{mut}}(t) & \text{orig}_t=C \ \text{and}\ O_t=T \
0 & \text{otherwise}
\end{cases}
]

This is a *very sharp* emission model: most mismatches are impossible in state 1.

## 6) The forward algorithm (log-likelihood)

### Purpose

Compute:
[
\log P(O_{0:T-1} \mid \text{orig}, \theta)
]
by summing over all hidden-state paths:
[
P(O) = \sum_{S_{0:T-1}} P(S_0)\prod_{t=1}^{T-1}P(S_t\mid S_{t-1}) \prod_{t=0}^{T-1}P(O_t\mid S_t)
]

### Transitions

Parameters:

* `alpha_recruit`: probability of going **0→1**
* `lambda_scan`: probability of staying **1→1**

So:

* From 0:

  * 0→1 = α
  * 0→0 = 1−α
* From 1:

  * 1→1 = λ
  * 1→0 = 1−λ

This is a 2-state Markov chain with “recruitment into scanning” and “persistence of scanning”.

### Log transition matrix

```py
log_trans = [[log(1-α), log(α)],
             [log(1-λ), log(λ)]]
```

### Initialization

```py
# Start forced in state 0 at t=0 (as in generator)
alpha[0,0] = log_emit_bg(...)
alpha[0,1] = -inf
```

HMM concept: this sets **initial distribution** π as:

* π0 = 1
* π1 = 0

So the model assumes AID begins unbound.

### Recurrence (core forward step)

For each t≥1:

State 0:

```py
alpha[t,0] = loge_bg + logsumexp([
    alpha[t-1,0] + log_trans[0,0],
    alpha[t-1,1] + log_trans[1,0]
])
```

This is:
[
\log \alpha_t(0)=\log P(O_{0:t}, S_t=0)
= \log P(O_t\mid S_t=0) + \log\left(\sum_{s'} \alpha_{t-1}(s')P(S_t=0\mid s')\right)
]

State 1:

```py
alpha[t,1] = loge_bound + logsumexp([
    alpha[t-1,0] + log_trans[0,1],
    alpha[t-1,1] + log_trans[1,1]
])
```

Same idea for state 1.

### Termination

```py
return logsumexp(alpha[T-1,:])
```

This returns:
[
\log P(O_{0:T-1})=\log\left(\alpha_{T-1}(0)+\alpha_{T-1}(1)\right)
]

### Why log-space + `logsumexp` matters

Raw probabilities multiply and underflow to zero for long sequences. Log-space turns products into sums, and `logsumexp` computes:
[
\log(e^a + e^b)
]
stably.

## 7) Dataset negative log-likelihood

```py
total -= ll
```

Summing log-likelihood across independent sequences corresponds to multiplying likelihoods:

[
\log P(\text{all data}) = \sum_i \log P(O^{(i)} \mid \theta)
]

Then the optimizer minimizes negative log-likelihood:

[
\text{NLL}(\theta) = -\sum_i \log P(O^{(i)}\mid \theta)
]

The `1e30` is just a big penalty if a parameter setting makes likelihood invalid.

## 9) How to map every parameter to HMM concepts

### Transition parameters

* `alpha_recruit` controls **expected entry rate** into state 1.

  * Larger α → more frequent bound episodes.
* `lambda_scan` controls **expected duration** of bound episodes.

  * Expected run length in state 1 is roughly (1/(1-\lambda)).

These are classic Markov-chain “switching” controls.

### Emission parameters (state 1 only)

* `base_activation`: baseline firing rate on C in bound state.
* `hotspot_bias`: multiplier for WRC contexts.
* `coldspot_bias`: multiplier for SYC contexts.
* `mutation_eff`: converts “firing” into *observable* C→T, i.e. repair failure.

In your synthetic story:

* state 1 = “enzyme bound and scanning”
* emission = “did it cause a C→T here or not”
* context controls the probability of seeing the mutation

### Fixed parameter

* `NOISE` is the state-0 emission error rate.

## 10) A few subtle but important model implications (mastery-level)

1. **State 1 forbids non-(C→T) mismatches.**
   If your mutated sequences contain A→G, C→A, etc., the model will be forced to explain them via **state 0 noise**, not bound state. That can strongly affect inferred α and λ.

2. **The model is “input-dependent HMM.”**
   Emissions depend on `(orig_seq, position)` not just state. This is still an HMM because the hidden-state process is Markov and emissions are conditionally independent given state + input.

3. **Starting in state 0 biases early positions.**
   Positions near the start can only be explained as background until the chain transitions 0→1.

4. **Clamp creates saturation effects.**
   If `base_activation * hotspot_bias ≥ 1`, then hotspot firing saturates at 1 and you lose identifiability between those two parameters in that regime (many combinations behave the same).

5. **This computes likelihood, but does not decode states.**
   You’re not running Viterbi or posterior decoding. If you want the inferred binding track along the sequence, you’d add:

   * Viterbi (most likely path)
   * or forward-backward (posterior (P(S_t=1\mid O)))

## 11) If you want to “associate it with HMM concepts” in one mental model

* The hidden state is a **latent annotation track** along DNA:

  * 0 = background
  * 1 = AID-bound
* The transition model makes state 1 appear in **contiguous runs** (bursts of scanning).
* The emission model says:

  * In state 0 you mostly copy the original with rare random errors.
  * In state 1, you copy the original except sometimes you produce a **very specific mutation** (C→T), and only at Cs, and more often in hotspots.

Everything else is just:

* **forward algorithm** to sum over all possible binding tracks
* **numerical optimization** to find the parameters that best explain observed mutations
